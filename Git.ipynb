```python
%matplotlib inline 
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np

plt.style.use('seaborn-whitegrid')# background color
# importing the data
data=pd.read_csv("heart-disease.csv")
fig, (ax1,ax2)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True);

# adding the data in ax1
s=ax1.scatter(x=data['age'],y=data['chol'],c=data['target'],cmap='plasma')
ax1.set(ylabel='cholesterol',title="Heart Disease")
ax1.legend(*s.legend_elements(),title='Target')
ax1.axhline(data['chol'].mean(),linestyle='--',color='r')
ax1.set_xlim(30,90)

# adding data in ax2
s2=ax2.scatter(x=data['age'],y=data['thalach'],c=data['target'],cmap='winter')
ax2.set(xlabel="Age",ylabel="Heart-Rate",title="Heart Disease")
ax2.legend(*s.legend_elements(),title='Target',edgecolor='r')
ax2.axhline(data['thalach'].mean(),linestyle='--',color='r')
ax2.set_ylim(50,220)

# Adding a title to whole figure

fig.suptitle("Heart Dieseas Analysis",fontsize=16,fontweight='bold');
```
